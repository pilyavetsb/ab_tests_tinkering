```python
import numpy as np
import names
import pandas as pd
import scipy.stats as stats
import plotly.express as px
from sklearn.model_selection import train_test_split
```

```python
s_dev = s = 1.21
rev = stats.lognorm.rvs(s_dev, size=5000, random_state = 7)
fig = px.histogram(rev)
fig.show()
```

```python
names_list = [names.get_full_name() for i in range(len(rev))]
traf_dict = {
    "revenue": rev,
    "name": names_list
}
df = pd.DataFrame.from_dict(traf_dict)
df.head()
```

```python
df['revenue'] *= 1000
```

```python
threshold = 0.5
labels = np.random.rand(len(df)) < 0.5
df['label'] = labels
```

```python
stats.ttest_ind(a=df.loc[df['label'], 'revenue'], b=df.loc[~df['label'], 'revenue'], equal_var=False).pvalue
```

```python
res = list()
for i in range(1000):
    labels = np.random.rand(len(df)) < threshold
    df['label'] = labels
    res.append(
        stats.ttest_ind(a=df.loc[df['label'], 'revenue'], b=df.loc[~df['label'], 'revenue'], equal_var=False).pvalue
        )
```

```python
fig1 = px.histogram(res)
fig1.show()
```
